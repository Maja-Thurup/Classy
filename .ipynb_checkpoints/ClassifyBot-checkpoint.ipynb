```python
#Introduction Function

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

def intro():
    print('––––––––––––––––––––––––––––––––––––')
    print(color.BOLD+'Welcome to CLASSIFY-BOT!'+color.END)
    print('––––––––––––––––––––––––––––––––––––\n')
```

```python
#Generic Error Message Function
def error():
    print(color.RED +'\nWhoops, something went wrong. Please try launching Classify-Bot again.'+color.END)
```

```python
#Get Files Function

import os
from os import walk

def get_files():
    mypath = os.path.abspath(os.getcwd())
    files = []
    for (dirpath, dirnames, filenames) in walk(mypath):
        files.extend(filenames)
        break
    csvs = []
    for file in files:
        if '.csv' in file:
            csvs.append(file)
    print('I found ' + color.BLUE + color.BOLD + str(len(csvs)) + color.END +' CSV files in your directory that we can get started with.')
    if len(csvs) > 0:
        print('Which data set do you want to work with?\n')
        i = 1
        for csv in csvs:
            print(i,'-',csv)
            i+=1
        print(color.GREEN + '\nEnter the number of the file'+ color.END)
        
        file_choice = input()
        
        if file_choice.isnumeric() and (int(file_choice) in range(1,len(csvs)+1)):
            return csvs[int(file_choice)-1]
        else:
            error()
            return(None)
    else:
        print('Please upload a CSV File to Begin')
        return(None)
```

```python
#Scrubbing Function

import pandas as pd
import numpy as np
import sys
def scrub(filename):   
    df = pd.read_csv(filename)
    scrubbed_df = pd.read_csv(filename)
    col = df.shape[1]
    row = df.shape[0]
    print('\n' + filename + 'has ' + str(col) + ' columns and ' + str(row) + ' rows')
    print("We'll start by going column by column to make sure everything's in order\n")
    
    i = 1 #counter
    includer = 0
    
    for col in df.columns:
        includer = 0
        print(color.BOLD + 'Column '+str(i) + '/'+str(len(df.columns))+':',col + color.END)
        print('Data Type:',df[col].dtype)
        if(df[col].dtype == np.float64 or df[col].dtype == np.int64):
            print('Mean:',round(df[col].mean(),2))
            print('Min:',round(df[col].min(),2))
            print('Max:',round(df[col].max(),2))
        print('Missing Data:',df[col].isna().sum())
        
        #Include or drop
        yorno = input('\nWould you like to include this column in the model?'+color.GREEN + ' Type (y/n) below'+ color.END+'\n')
        if yorno == 'y':
            print('Column was successfully added.\n')
            includer = 1
        elif yorno == 'n':
            print('Okay, the column was removed.\n')
            i+=1
            scrubbed_df = scrubbed_df.drop(col,axis=1)
        else:
            error()
            i+=1
            return None
      
        if includer == 1:
            #Ensure correct datatype
            if(df[col].dtype == np.float64 or df[col].dtype == np.int64):
                yorno = input('\nIs this column a category? / is the data type right.'+color.GREEN + ' Type (y/n) below'+ color.END+'\n')
                if yorno == 'y':
                    print('Great, no changes necessary then.\n')
                elif yorno == 'n':
                    print("Okay, I'll change it to a category.\n")
                    scrubbed_df[col] = scrubbed_df[col].astype(str)
                else:
                    error()
                    i+=1
                    return None

            #Deal with missing data
            if (df[col].isna().sum() != 0):
                print('We have to deal with some missing data.')
            i+=1
    return scrubbed_df   
```

```python
def classifybot():
    intro()

    my_file = get_files()
    if my_file == None:
        return
    scrubbed_df = scrub(my_file)
    print(scrubbed_df.columns)
```

```python
classifybot()
```
